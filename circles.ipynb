```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
img = np.ones((1200 , 1200 , 3), dtype = np.uint8) * 255
r , c , _ = img.shape 

center = (600 , 600)
number_of_circles = 25

for i in range(number_of_circles):
    radius = (number_of_circles - i) * 25
    color = (255 , 255 , 0)
    thickness = i + 1
    
    cv2.circle(img , center, radius , color , thickness)
    


plt.imshow(img)
plt.axis('off');
```
