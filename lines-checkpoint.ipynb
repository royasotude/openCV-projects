### Imports

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

### drawing lines

```python
img = np.zeros((1200, 1200 , 3) , dtype= np.uint8)
r , c , _= img.shape

for i in range(0 , r-1 , 99):

     pt1 = (i , 0)
     pt2 = (r-1 , i)
     color = (255 , 105 , 180)
     thickness = 4
     cv2.line(img , pt1 , pt2 , color , thickness)


     pt1 = (r-i, 0)
     pt2 = (0, i)
     color = (255, 215, 0)
     thickness = 4
     cv2.line(img, pt1, pt2, color, thickness)


     pt1 = (0, i)
     pt2 = (i, r-1)
     color = (0, 0, 128)
     thickness = 4
     cv2.line(img, pt1, pt2, color, thickness)


     pt1 = (i, r-1)
     pt2 = (r-1, r-i)
     color = (0, 255, 255)
     thickness = 4
     cv2.line(img, pt1, pt2, color, thickness)



plt.imshow(img)
plt.axis('off');
```
